```python
from transformers import pipeline

# Usa un modello BERT pre-addestrato fine-tuned per question answering
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Il contesto o passaggio di testo in cui BERT deve trovare la risposta
context = """
I hate the war in Ukraine, I hope it ends as soon as possible!.
"""

# La domanda
question = "What feeling does your context express between (negative), (positive) and (neutral)? Answer only with the answers in brackets"

# Usa BERT per trovare la risposta nel contesto
response = qa_pipeline(question=question, context=context)

# Stampa la risposta
print(response['answer'])
```
